```
import pandas as pd
import numpy as np
import math
from sklearn.linear_model import LinearRegression
```

```
import pandas as pd

data = pd.read_csv('/content/thucnghiem.txt')
print(data)

import matplotlib.pyplot as plt

data.columns = ['size', 'operations']
import matplotlib.pyplot as plt
plt.scatter(data['size'], data['operations'])
```

##Chuẩn hóa dữ liệu


```
X = data['size'].values.reshape(-1,1)
```

##Hàm tìm hệ số X

```
def Find_X(_X):
  regr = LinearRegression()
  regr.fit(_X, data['operations'])
  return regr.coef_
```

```
result = []
```

##Hàm lg(n)

```
X_ = np.log2(X)
re = Find_X(X_)
print(re)
result.append(re)
```

##Hàm sqrt

```
X_ = np.sqrt(X)
re = Find_X(X_)
print(re)
result.append(re)
```

##Hàm N

```
re = Find_X(X)
print(re)
result.append(re)
```

##Hàm n*log(n)

```
X_ = X*np.log2(X)
re = Find_X(X_)
print(re)
result.append(re)
```

##Hàm pow(n,2)

```
X_ = X**2
re = Find_X(X_)
print(re)
result.append(re)
```

##Hàm pow(n,3)

```
X_ = X**3
re = Find_X(X_)
print(re)
result.append(re)
```

##Hàm pow(2,n)

```
X_ = 2**X
re = Find_X(X_)
print(re)
result.append(re)
```

```
print(result)
```
